## Mini práctica A: Registro de gastos

Uso de listas, diccionarios y ciclos para calcular gastos por categoría.

```python
gastos = [
    {"Categoria": "Comida", "Monto": 1200},
    {"Categoria": "Transporte", "Monto": 500},
    {"Categoria": "Herramientas", "Monto": 1000},
    {"Categoria": "Entretenimiento", "Monto": 100}
]

total = 0

for gasto in gastos:

    categoria = gasto["Categoria"]
    montos = gasto["Monto"]

    if type(montos) != list:
      montos = [montos]

    total_categoria = sum(montos)
    print(f"En {categoria} la cantidad fue de: {total_categoria}")
    total += total_categoria

print()
print("El monto total de todas las categorias es", total)
```

Con varios montos ahora:

```python
gastos = [
    {"Categoria": "Comida", "Monto": [1200,1000,500]},
    {"Categoria": "Transporte", "Monto": [500,250]},
    {"Categoria": "Herramientas", "Monto": [1000,500]},
    {"Categoria": "Entretenimiento", "Monto": [100,300]}
]

total = 0

for gasto in gastos:

    categoria = gasto["Categoria"]
    montos = gasto["Monto"]
    total_categoria = sum(montos)
    print(f"En {categoria} la cantidad fue de: {total_categoria}")
    total += total_categoria

print()
print("El monto total de todas las categorias es", total)
```

## Mini práctica B: Perfil de usuario

Uso de variables, entrada de datos e impresión de información básica.

```python
perfil = {}

perfil["Nombre"] = input("Ingresa tu nombre: ")

perfil["Edad"] = int(input("Ingresa tu edad: "))

perfil["Ciudad"] = input("Ingresa tu ciudad: ")

perfil["Hobbies"] = input("Ingresa tus hobbies (separados por comas): ").split(",")

print("-"*80)
print()
print("Perfil del usuario:")
print()

for clave, valor in perfil.items():
    print(clave, ":", valor)
```
